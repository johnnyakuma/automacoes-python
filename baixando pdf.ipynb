```python
from selenium import webdriver
import time
import requests
import re
import PyPDF2

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# Inicializa uma instância do navegador Chrome e a atribui à variável 'navegadorFormulario'
navegadorFormulario = webdriver.Chrome()

# Navega para a URL especificada.
navegadorFormulario.get('https://www.iades.com.br/inscricao')

time.sleep(1)

# Seleciona o campo mencionado na variável com XPATH
nome = navegadorFormulario.find_element(By.XPATH, '//*[@id="botaoConcurso"]/a/img')
# Usa a função para selecionar o elemento e clicar como um humano
nome.click()

time.sleep(1)

# Localiza o elemento do link para o PDF
pdf = navegadorFormulario.find_element(By.XPATH, '//*[@id="fieldset1"]/table/tbody/tr/td[2]/a')

# Extrai o URL do link
url_pdf = pdf.get_attribute('href')

# Baixa o conteúdo do PDF usando requests
response = requests.get(url_pdf)
if response.status_code == 200:
    # Especifica o caminho onde o PDF será salvo localmente
    file_path = "exemplo.pdf"
    # Abre o arquivo local em modo de escrita binária
    with open(file_path, 'wb') as f:
        # Escreve o conteúdo do PDF no arquivo local
        f.write(response.content)
    print("Download do PDF concluído com sucesso!")

    # Abre o arquivo PDF baixado e extrai o texto
    with open(file_path, 'rb') as file:
        leitor_pdf = PyPDF2.PdfReader(file)
        texto_pdf = ""
        for pagina_num in range(len(leitor_pdf.pages)):
            pagina = leitor_pdf.pages[pagina_num]
            texto_pdf += pagina.extract_text()

    # Padrao regex para encontrar valores monetarios
    padrao = r'R\$\s*\d{1,3}(?:\.\d{3})*(?:,\d{2})?'

    # Encontra todas as ocorrências do padrão no texto
    valores_monetarios = re.findall(padrao, texto_pdf)

    # Exibe os valores encontrados
    print("Valores monetários encontrados:")
    for valor in valores_monetarios:
        print(valor)
else:
    print("Erro ao baixar o PDF:", response.status_code)

# Fecha o navegador
navegadorFormulario.quit()

print('Automação concluída!')
```
