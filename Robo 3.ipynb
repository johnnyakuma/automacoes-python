```python
from openpyxl import load_workbook
import os

#pegamos o caminho + nome do arquivo
nome_arquivo_cep = 'C:\\Users\\johnn\\Documents\\Curso RPA\\robo2 excels salvos\\PesquisaEndereco_2.xlsx'
planilha_dados_endereco = load_workbook(nome_arquivo_cep)

#selecionando a sheet de dados
sheet_selecionada = planilha_dados_endereco["CEP"]


from selenium import webdriver as opcoes_selenium
from selenium.webdriver.common.keys import Keys
import pyautogui as tempo_espera
from selenium.webdriver.common.by import By

meuNavegador = opcoes_selenium.Chrome()
#abre o google chrome no site do busca cep
meuNavegador.get('https://buscacepinter.correios.com.br/app/endereco/index.php')

#tempo de espera
tempo_espera.sleep(3)

#procurando pelo elemento NAME e quando encontrador escrever a SEND_KEY OBS: apenas teste 
meuNavegador.find_element(By.NAME, 'endereco').send_keys('05892-387')

tempo_espera.sleep(2)

#clicando no botao Buscar
meuNavegador.find_element(By.NAME, 'btn_pesquisar').click()

#tempo de espera
tempo_espera.sleep(2)


for linha in range (2, len(sheet_selecionada['A']) + 1):
    
    #tempo de espera
    tempo_espera.sleep(2)
    #pesquisando pelo id para clicar para voltar
    meuNavegador.find_element(By.ID, 'btn_nbusca').click()
    
    tempo_espera.sleep(2)
    
    #pegando o cep da planilha e colocando na variavel
    cepPesquisa = sheet_selecionada['A%s'% linha].value
    
    tempo_espera.sleep(2)
    
    #procurando pelo elemento NAME e quando encontrador escrever a SEND_KEY
    meuNavegador.find_element(By.NAME, 'endereco').send_keys(cepPesquisa)

    tempo_espera.sleep(2)

    #clicando no botao Buscar
    meuNavegador.find_element(By.NAME, 'btn_pesquisar').click()

    #tempo de espera
    tempo_espera.sleep(2)
    
    #pega os dados da variavel mencionada
    rua = meuNavegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[1]')[0].text
    print(rua)
    #pega os dados da variavel mencionada
    bairro = meuNavegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[2]')[0].text
    print(bairro)
    #pega os dados da variavel mencionada
    cidade = meuNavegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[3]')[0].text
    print(cidade)
    #pega os dados da variavel mencionada
    cep = meuNavegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[4]')[0].text
    print(cep)
    
    #seleciona a sheet de dados
    sheet_dados_para_imprimir_endereco = planilha_dados_endereco["Dados"]
    linhaCorrentePlanilhaCEP = len(sheet_dados_para_imprimir_endereco['A']) + 1
    
    #pegamos a ultima linha preenchida na coluna A e acrescentamos +1
    coluna_A = "A" + str(linhaCorrentePlanilhaCEP) #criando a variavel para juntar A + linha => ex:A2
    coluna_B = "B" + str(linhaCorrentePlanilhaCEP) #criando a variavel para juntar B + linha => ex:B2
    coluna_C = "C" + str(linhaCorrentePlanilhaCEP) #criando a variavel para juntar C + linha => ex:C2
    coluna_D = "D" + str(linhaCorrentePlanilhaCEP) #criando a variavel para juntar D + linha => ex:D2

    #imprimindo as informacoes do site na planilha excel
    sheet_dados_para_imprimir_endereco[coluna_A] = rua #A2 A3 A4...
    sheet_dados_para_imprimir_endereco[coluna_B] = bairro #B2 B3 B4...
    sheet_dados_para_imprimir_endereco[coluna_C] = cidade #C2 C3 C4...
    sheet_dados_para_imprimir_endereco[coluna_D] = cep #D2 D3 D4...

#salvando o arquivo do excel com as novas informações
planilha_dados_endereco.save(filename=nome_arquivo_cep)

#abrindo e apresentando na tela o arquivo com os dados
os.startfile(nome_arquivo_cep)
```
